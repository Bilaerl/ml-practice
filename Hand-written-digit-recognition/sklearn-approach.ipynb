Starting with a simple model before trying more complex ones, we try to tackle the problem using a Logistic Regression model

```python
#all necessary imports
import sklearn.datasets as datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function to show image and it's label
def show_image(image, true_label, pred_label=None):
    image = image.reshape(8,8)
    plt.imshow(image, cmap="Greys")
    
    title = f"True label: {true_label}"
    if pred_label or pred_label == 0:
        title += f" \n Predicted as: {pred_label}"
    
    plt.title(title)
```

```python
X, y = datasets.load_digits(return_X_y=True) # load data
print(f"Shape of X: {X.shape} \t Shape of y: {y.shape}") # print shape of data
```

```python
# split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8, random_state=42)
print(f"X_train: {X_train.shape} \t y_train: {y_train.shape}") # print shape of train data
print(f"X_test: {X_test.shape} \t y_test: {y_test.shape}") # print shape of test data
```

```python
# print one of our train data to make sure data is ok
# you can change the index and print if you want to see more examples
index = 12
show_image(X_train[index], y_train[index])
```

```python
# create a logistic regression model
log_reg_model = LogisticRegression(random_state=42, solver='newton-cg')

# fit model to train data
log_reg_model.fit(X_train,y_train)
```

```python
log_reg_model.score(X_train,y_train) # accuracy of model on train data
```

```python
log_reg_model.score(X_test, y_test) # Accuracy of model on test data
```

```python
preds = log_reg_model.predict(X_test) # make prediction on test data
```

```python
# show image with true label and predicted label to confirm model accuraxy
# you can change the index and print if you want to see more examples
index = 13
show_image(X_test[index], y_test[index], preds[index])
```

```python
# visualise examples that the model got wrong
for image, true, pred in zip(X_test, y_test, preds):
    if true != pred:
        plt.figure()
        show_image(image, true, pred)
```
